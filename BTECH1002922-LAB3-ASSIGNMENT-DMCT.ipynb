#### QUESTION 1

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
 
 
data = {
    'COUNTRY': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'SOLD PRICE': [50000, 450000, 600000, 550000, 470000]
}
DF = pd.DataFrame(data)
 
 
plt.figure(figsize=(10, 6)) 
sns.barplot(data=DF, x='COUNTRY', y='SOLD PRICE')
 
 
plt.title('Sold Price by Country')
plt.xlabel('Country')
plt.ylabel('Sold Price')
 
 
plt.show()
```

#### INTERPRETATION:
This bar chart shows the average SOLD PRICE for each COUNTRY. It helps in understanding which countries have higher or lower average SOLD PRICE values.

#### RESULT: 
UK has the highest Sold Price amongst all the countries.

#### QUESTION 5

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
 
 
data = {
    'SOLD PRICE': [500000, 450000, 600000, 550000, 470000, 650000, 480000, 530000, 490000, 540000]
}
DF = pd.DataFrame(data)
 
 
Q1 = DF['SOLD PRICE'].quantile(0.25)
Q3 = DF['SOLD PRICE'].quantile(0.75)
IQR = Q3 - Q1
 
 
outliers = DF[(DF['SOLD PRICE'] < (Q1 - 1.5 * IQR)) | (DF['SOLD PRICE'] > (Q3 + 1.5 * IQR))]
 
 
plt.figure(figsize=(10, 6))
sns.boxplot(x=DF['SOLD PRICE'])
 
 
plt.title('Boxplot of Sold Prices')
plt.xlabel('Sold Price')
 
plt.show()
 
 
print("Outliers:\n", outliers)
```

#### INTERPRETATION:
The boxplot illustrates the median, quartiles, and outliers in SOLD PRICE data. Outliers are identified as points outside the whiskers of the box.

#### QUESTION 10

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
 
 
data = {
    'SR-B': [1.2, 2.3, 1.8, 2.5, 3.0, 2.7, 2.0, 1.5, 2.4, 1.9],
    'AVE': [25.5, 30.2, 27.1, 29.8, 31.5, 28.9, 26.3, 24.8, 29.0, 27.6],
    'SIXERS': [12, 8, 10, 5, 7, 9, 11, 6, 13, 10],
    'SOLD PRICE': [500000, 450000, 600000, 550000, 470000, 650000, 480000, 530000, 490000, 540000]
}
DF = pd.DataFrame(data)
 
 
features = ['SR-B', 'AVE', 'SIXERS', 'SOLD PRICE']
 
 
correlation_matrix = DF[features].corr()
 
 
plt.figure(figsize=(10, 8)) 
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
 
 
plt.title('Correlation Matrix of Features')
 
 
plt.show()
```

#### INTERPRETATION:
The correlation matrix heatmap shows the correlation coefficients between pairs of features, highlighting strong or weak correlations.
